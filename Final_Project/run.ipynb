```python
import pandas as pd

# بارگذاری داده‌های نودها
nodes_df = pd.read_csv('./telegram/telegram_10_to_24_dey_node.csv')

# مرتب‌سازی داده‌ها برای هر فیلد و انتخاب 5 نود برتر
top_posts = nodes_df[['nodedef_name', 'posts_INT']].sort_values(by='posts_INT', ascending=False).head(5)
top_followers = nodes_df[['nodedef_name', 'followers_INT']].sort_values(by='followers_INT', ascending=False).head(5)
top_views = nodes_df[['nodedef_name', 'views_INT']].sort_values(by='views_INT', ascending=False).head(5)

print(top_posts)
print(top_followers)
print(top_views)
```

```python
import networkx as nx

# بارگذاری داده‌های یال‌ها
edges_df = pd.read_csv('./telegram/telegram_10_to_24_dey_edge.csv')

# ساخت گراف
G = nx.DiGraph()
for _, row in edges_df.iterrows():
    G.add_edge(row['node1'], row['node2'], weight=row['weight_DOUBLE'])

# محاسبه Hub, Authority و PageRank
hub, authority = nx.hits(G)
pagerank = nx.pagerank(G)

# استخراج 10 نود برتر برای هر معیار
top_hub = sorted(hub.items(), key=lambda x: x[1], reverse=True)[:10]
top_authority = sorted(authority.items(), key=lambda x: x[1], reverse=True)[:10]
top_pagerank = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)[:10]

print("Top 10 Hubs:", top_hub)
print("Top 10 Authorities:", top_authority)
print("Top 10 PageRank:", top_pagerank)
```

```python
# حذف جهت یال‌ها
G_undirected = G.to_undirected()

# اجرای الگوریتم k-core
k_core = nx.k_core(G_undirected)

# محاسبه k-shell
k_shell = nx.core_number(G_undirected)

# مشاهده نودهای با بیشترین k-shell
top_k_shell = sorted(k_shell.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top k-shell nodes:", top_k_shell)
```
